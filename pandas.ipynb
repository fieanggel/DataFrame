```python
import pandas as pd

data = pd.read_excel('disperkim-od_16985_jumlah_produksi_sampah_berdasarkan_kabupatenkota_v3_data.xlsx', sheet_name='data')
data 
```

```python
sampahtotal = 0
tahun = 2019
for index, row in data.iterrows():
    if row['tahun']== tahun:
        sampahtotal += row ['jumlah_produksi_sampah']
print(f"Total sampah tahun 2019: {sampahtotal} Ton")
```

```python
data = pd.DataFrame([{'Tahun': tahun, 'Total sampah (Ton)': sampahtotal}])
data
# data.to_csv('no2.csv', index=False)
# data.to_excel('no2.xlsx', index=False)
```

```python
smph_per_tahun = {}

for index, row in data.iterrows():
    tahun = row['tahun']
    jumlah = row['jumlah_produksi_sampah']
    if tahun in smph_per_tahun:
        smph_per_tahun[tahun] += jumlah
    else:
        smph_per_tahun[tahun] = jumlah

data = pd.DataFrame(smph_per_tahun.items(), columns=['Tahun', 'Total Produksi Sampah'])
data
# data.to_csv('no3.csv', index=False)
# data.to_excel('no3.xlsx', index=False)
```

```python
jml_data_perkota = {}

for index, row in data.iterrows():
    kota = row['nama_kabupaten_kota']
    tahun = row['tahun']
    jumlah = row['jumlah_produksi_sampah']
    
    if kota not in jml_data_perkota:
        jml_data_perkota[kota] = {}
    
    if tahun not in jml_data_perkota[kota]:
        jml_data_perkota[kota][tahun] = 0
    
    jml_data_perkota[kota][tahun] += jumlah

data = pd.DataFrame([
    {'Kabupaten/Kota': kota, 'Tahun': tahun, 'Jumlah Data': jumlah}
    for kota, data_tahun in jml_data_perkota.items()
    for tahun, jumlah in data_tahun.items()
])
data
data.to_csv('No4.csv', index=False)
data.to_excel('No4.xlsx', index=False)
```
